<b>Analisis Dampak Penggunaan Media Sosial Pada Anak Muda: Kesehatan Mental, Produktivitas, dan Pola Konsumsi</b>
<ol>
    <li>id responden</li>
    <li>usia : 12 - 24 </li>
    <li>jenis kelamin :  L/P -> L(1), P(0)</li>
    <li>paltform yang disukai : instagram, tiktok, youtube, x, facebook</li>
    <li>durasi penggunaan : jam/hari</li>
    <li>tipe konten : Hiburan, Edukasi, Berita, Viral Challenge</li>
    <li>produktivitas : skala 1-10, self-assessment </li>
    <li>kesehatan mental : skala 1-10, berdasarkan kuesioner stres/kecemasan</li>
    <li>oversharing : postingan data pribadi (sering/jarang/tidak pernah)</li>
    <li>dampak : positif/netral/negatif</li>
</ol>

```python
import numpy as np
import pandas as pd

n_data = 1000

id_responden = np.arange(1, n_data + 1)
usia = np.random.randint(12,25, size = n_data)
jenis_kelamin = np.random.randint (0,2, size = n_data)
platform = np.random.choice(['Instagram', 'TikTok', 'YouTube', 'X', 'Facebook'], size = n_data, p = [0.3, 0.4, 0.2, 0.08, 0.02])
durasi_penggunaan = np.round(np.random.normal(loc = 4, scale = 1.5, size = n_data), 1)
durasi_penggunaan = np.clip(durasi_penggunaan, 1, 8)
konten = np.random.choice(['Hiburan', 'Edukasi', 'Berita', 'Viral Challenge'], size = n_data, p = [0.6, 0.25, 0.1, 0.05])
produktivitas = np.random.randint(2, 11, size = n_data)
kesehatan_mental = np.clip(10-(durasi_penggunaan * 0.7 + np.random.normal(0,1, n_data)),1,10).astype(int)
oversharing = np.random.choice(['sering', 'jarang', 'tidak pernah'], size = n_data, p = [0.2, 0.5, 0.3])
dampak = np.where((durasi_penggunaan > 5) | (kesehatan_mental < 5), 'negatif', np.where(durasi_penggunaan < 3, 'posistif', 'netral'))

df = pd.DataFrame({
    'id_responden' : id_responden,
    'usia' : usia,
    'jenis_kelamin' : jenis_kelamin,
    'platform' : platform,
    'durasi_penggunaan (jam/hari)' : durasi_penggunaan,
    'konten' : konten,
    'produktivitas' : produktivitas,
    'kesehatan_mental' : kesehatan_mental,
    'oversharing' : oversharing,
    'dampak' : dampak
})

print(df)
print(df.info())
print(df.head())
print(df.describe())
print(f'usia termuda :  {df['usia'].min()} tahun')
print(f'usia tertua : {df['usia'].max()} tahun')
print(f'rata-rata durasi penggunaan : {df['durasi_penggunaan (jam/hari)'].mean():2f} jam/hari')
jmlL = df[df['jenis_kelamin']==1]
jmlp = df[df['jenis_kelamin']==0]
print(jmlL['jenis_kelamin'].size)
print(jmlp['jenis_kelamin'].size)
df.to_csv('TUGAS 10_dataPenggunaaMedsos.csv')
```
